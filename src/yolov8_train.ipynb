## Visualize some predictions

```python
def visualize_segmentation(results):
    # Extract the original image
    img = results.orig_img
    orig_height, orig_width = img.shape[:2]

    # Check if masks are present in the results
    if results.masks is not None:
        masks = results.masks.data.cpu().numpy()  # Convert masks to a NumPy array
        boxes = results.boxes.xyxy.cpu().numpy()  # Bounding boxes
        class_ids = results.boxes.cls.cpu().numpy().astype(int)  # Class IDs

        # Resize each mask to match the original image size
        for i, mask in enumerate(masks):
            mask = cv2.resize(mask, (orig_width, orig_height), interpolation=cv2.INTER_NEAREST).astype(np.uint8)
            color = np.random.randint(0, 255, (3,), dtype=np.uint8)  # Random color for the mask
            colored_mask = np.zeros_like(img, dtype=np.uint8)
            for c in range(3):
                colored_mask[:, :, c] = mask * color[c]
            img = cv2.addWeighted(img, 1, colored_mask, 0.5, 0)  # Blend mask with the image

            # Add bounding box and class label
            x1, y1, x2, y2 = boxes[i]
            label = f"{results.names[class_ids[i]]}"
            cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), color.tolist(), 2)
            cv2.putText(img, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_PLAIN, 0.9, color.tolist(), 2)

        # Display the image with masks
        plt.figure(figsize=(10, 10))
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()
    else:
        print("No masks were predicted.")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import os
import random
import cv2
%matplotlib inline

data_path = 'D:/UoL/Final Project/src/datasets/yolo'  # Your dataset directory

validation_images_dir = os.path.join(data_path, 'test', 'images')
image_files = [f for f in os.listdir(validation_images_dir)]

# Randomly select images
selected_images = random.sample(image_files, 5)

for img_file in selected_images:
    img_path = os.path.join(validation_images_dir, img_file)
    results = model.predict(img_path)[0]  # Replace with your image path
    visualize_segmentation(results)

```
